## PMM pricing formula

### 对于 Base

$$P = i (1 - k + k(\frac{B_0}{B})^2), \{k \in R : 0 \leq k \leq 1\}$$

### 对于 Quote

$$P = i / (1 - k + k (\frac{Q_0}{Q})^2), \{k \in R : 0 \leq k \leq 1\}$$

- $i$: mid price
- $B$: 当前 Base Token 库存
- $B_0$: 初始 Base Token 库存
- $Q$: 当前 Quote Token 库存
- $Q_0$: 初始 Quote Token 库存

#### 当 k = 0, 价格不变

```python
import numpy as np 
from matplotlib import pyplot
k = 0
i = 4000    # mid price
B0 = 100    # initial base token stock
B = np.arange(30, B0)   # base token stock
P =  i * (1 - k + k * ((B0 / B) ** 2))
pyplot.title("PMM pricing curve (k = 0)") 
pyplot.xlabel("ETH stock") 
pyplot.ylabel("ETH Price") 
pyplot.plot(B, P) 
pyplot.show()
```

#### 当 k = 1 时，价格曲线正好是 AMM 中定义的 bonding curve:

```python
import numpy as np 
from matplotlib import pyplot
k = 1
i = 4000    # mid price
B0 = 100    # initial base token stock
B = np.arange(30, B0)   # base token stock
P =  i * (1 - k + k * ((B0 / B) ** 2))
pyplot.title("PMM pricing curve (k = 1)") 
pyplot.xlabel("ETH stock") 
pyplot.ylabel("ETH Price") 
pyplot.plot(B, P) 
pyplot.show()
```

#### k 越大，价格变化越敏感。k 越小，价格变化越不敏感

```python
import numpy as np 
from matplotlib import pyplot
k = 0.85
i = 4000    # mid price
B0 = 100    # initial base token stock
B = np.arange(30, B0)   # base token stock
P =  i * (1 - k + k * ((B0 / B) ** 2))
pyplot.title("PMM pricing curve (k = 0.85)") 
pyplot.xlabel("Remaining ETH stock") 
pyplot.ylabel("ETH Price") 
pyplot.plot(B, P) 
pyplot.show()
```
